```python
import numpy as np
import matplotlib.pyplot as plt
n_states = 5
alpha = 0.01
gamma = 1.0
episodes = 100
print("Divit shah || 221A046||38")
```

```python
def state_features(state):
    feature = np.zeros(n_states)
    # Check if 'states' is a valid 1-based index within the range [1, n_states]
    if 1 <= state <= n_states:
        feature[state-1] = 1
    return feature
# True values for plotting (from Sutton and Barto textbook)
true_values = np.array([1/6, 2/6, 3/6, 4/6, 5/6])

w = np.zeros(n_states)
print("Divit shah || 221A046||38")
```

```python
for ep in range(episodes):
    state = 3

    while True:
        next_state = state + np.random.choice([-1, 1])

        if next_state == 6:
            reward = 1.0
        else:
            reward = 0.0

        x = state_features(state)
        x_next = state_features(next_state)

        v = np.dot(w, x)
        v_next = np.dot(w, x_next) if 0 < next_state < 6 else 0

        td_error = reward + gamma * v_next - v
        w += alpha * td_error * x

        if next_state == 0 or next_state == 6:
            break

        state = next_state

estimated_values = np.dot(np.eye(n_states), w)
print("Divit shah || 221A046||38")
```

```python
print("Divit shah || 221A046||38")
plt.plot(true_values, label='True Values')
plt.plot(estimated_values, label='Estimated Values (Linear FA)')
plt.xlabel('State')
plt.ylabel('Value')
plt.legend()
plt.title('TD(0)with Linear Function Approximation')
plt.grid()
plt.show()
```
